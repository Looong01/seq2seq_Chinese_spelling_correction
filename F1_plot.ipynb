```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import config
import os
fig = plt.figure(figsize=(10, 10), dpi=300)
ax = fig.add_subplot(111,projection='3d')

# Grab some test data.
F1_path = os.path.join(config.results_path, 'F1.txt')
with open(F1_path,'r') as f:
    data = f.readlines()
batchsize = np.array([32, 64, 128, 256, 512, 1024])
dropout = np.array([0.05, 0.1, 0.15, 0.2, 0.25, 0.3])
F1 = np.array([float(c.strip()) for c in data])
x_z = [min(batchsize)-0.5]*len(batchsize)
y_z = [max(dropout)+0.5]*len(dropout)
x_y = [min(F1)-0.5]*len(F1)

ax.plot(x_z, dropout, F1)
ax.plot(batchsize, y_z, F1)
ax.plot(batchsize, dropout, x_y)
ax.set_xlabel('Batch Size')
ax.set_ylabel('Dropout')
ax.set_zlabel('F1')
ax.plot3D(batchsize,dropout,F1,'red')    #绘制空间曲线
plt.show()
# plt.savefig('F1.png')
```
